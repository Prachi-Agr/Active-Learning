```
df = pd.read_csv("./wine-clustering.csv")
df = df[["Alcohol","Malic_Acid", "Ash","Ash_Alcanity", "Magnesium",
         "Total_Phenols", "Flavanoids","Nonflavanoid_Phenols", "Proanthocyanins", "Color_Intensity",
         "Hue", "OD280", "Proline"]]
clustering_vars = ["Alcohol","Malic_Acid", "Ash","Ash_Alcanity", "Magnesium",
         "Total_Phenols", "Flavanoids","Nonflavanoid_Phenols", "Proanthocyanins", "Color_Intensity",
         "Hue", "OD280", "Proline"]
df = df.fillna(0)
data = df[clustering_vars].values
names = clustering_vars
```

```
net = sps.somNet(20, 20,data, PBC=True)
```

```
net.train(0.01, 10000)
```

```
net.save('wine_weights')
```

```
net.nodes_graph(colnum=0)
net.diff_graph()
```

```
net.project(data,labels=names)
```

```
net.cluster(data, type='qthresh')	
```
